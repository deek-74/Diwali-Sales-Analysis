```python
df.dropna(inplace=True)   #inplace likhna is like saving that deletion step, otherwise wapas show karne par null value display me aayegi, if inplace is not used. 
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Gender':'Sex'})   #save nahi hoga, to save use inplace operator
```

```python
df.describe()
```

```python
df[['Age', 'Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender', data=df)
```

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
sns.countplot(data=df, x='Age Group', hue='Gender')
```

```python
ax=sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)   #head is for specifying the number of states we want to see
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

                                                  Comparison between states(top 10) and Orders

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)   #head is for specifying the number of states we want to see
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state, x='State', y='Amount')
```

                                      Comparison between states and Amount

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(5,1)})
for bars in ax.containers:
    ax.bar_label(bars)
```
